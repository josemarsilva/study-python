## 3. Hands-on

```python
# Step-01: Read data/iris.data

with open('sample_data/iris.data') as file:
  lista = file.read().splitlines()
  print('n-length: ', len(lista))
  print('  0: ', lista[0])
  print('  1: ', lista[1])
  print('  2: ', lista[2])
  print('  :    :')
  print('%s: %s' % ( len(lista)-1, lista[len(lista)-3] ))
  print('%s: %s' % ( len(lista)-2, lista[len(lista)-2] ))
  print('%s: %s' % ( len(lista)-1, lista[len(lista)-1] )) 
```

```python
# Step-02: Clean Dataset last line 150: with empty contents
print('len(lista): ', len(lista))
lista.pop()
print('len(lista): ', len(lista))
```

```python
# Step-03: Identify distincts categories (classes)
lista_distinct_class = []
for item_lista in lista:
  if item_lista.split(',')[4]  not in lista_distinct_class:
    lista_distinct_class.append(item_lista.split(',')[4])

print('lista_distinct_class: ', lista_distinct_class)
```

```python
# Step-06: Identify: ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica'] x [ min(), max() ] x [ 'sepal', 'petal' ]  x [ width, length ]
# setosa
min_sepal_length = None
min_sepal_width = None
min_petal_length = None
min_petal_width = None
max_sepal_length = None
max_sepal_width = None
max_petal_length = None
max_petal_width = None
sum_sepal_length = 0
sum_sepal_width = 0
sum_petal_length = 0
sum_petal_width = 0

# 1. sepal length in cm, 2. sepal width in cm, 3. petal length in cm, 4. petal width in cm
for item_lista in lista:
  # default: min(), max()
  if min_sepal_length is None:
    min_sepal_length = float(item_lista.split(',')[0])
  if min_sepal_width is None:
    min_sepal_width = float(item_lista.split(',')[1]) 
  if min_petal_length is None:
    min_petal_length = float(item_lista.split(',')[2])
  if min_petal_width is None:
    min_petal_width = float(item_lista.split(',')[3]) 
  if max_sepal_length is None:
    max_sepal_length = float(item_lista.split(',')[0])
  if max_sepal_width is None:
    max_sepal_width = float(item_lista.split(',')[1]) 
  if max_petal_length is None:
    max_petal_length = float(item_lista.split(',')[2])
  if max_petal_width is None:
    max_petal_width = float(item_lista.split(',')[3]) 
  
  # Sum
  sum_sepal_length = sum_sepal_length + float(item_lista.split(',')[0])
  sum_sepal_width = sum_sepal_width + float(item_lista.split(',')[1]) 
  sum_petal_length = sum_petal_length + float(item_lista.split(',')[2])
  sum_petal_width = sum_petal_width + float(item_lista.split(',')[3]) 

  # min
  if float(item_lista.split(',')[0]) < min_sepal_length:
      min_sepal_length = float(item_lista.split(',')[0])
  if float(item_lista.split(',')[1]) < min_sepal_width:
      min_sepal_width = float(item_lista.split(',')[1])
  if float(item_lista.split(',')[2]) < min_petal_length:
      min_petal_length = float(item_lista.split(',')[2])
  if float(item_lista.split(',')[3]) < min_petal_width:
      min_petal_width = float(item_lista.split(',')[3])
  # max
  if float(item_lista.split(',')[0]) > max_sepal_length:
      max_sepal_length = float(item_lista.split(',')[0])
  if float(item_lista.split(',')[1]) > max_sepal_width:
      max_sepal_width = float(item_lista.split(',')[1])
  if float(item_lista.split(',')[2]) > max_petal_length:
      max_petal_length = float(item_lista.split(',')[2])
  if float(item_lista.split(',')[3]) > max_petal_width:
      max_petal_width = float(item_lista.split(',')[3])


print(f'')
print(f'+-------------------+-------+-------+------+---------------------+')
print(f'| Class             |  Min  |  Max  | Mean |                     |')
print(f'+-------------------+-------+-------+------+---------------------+')
print(f'| + sepal           |       |       |      |                     |')
print(f'|   - length        | {min_sepal_length:2.2f}  | {max_sepal_length:.2f}  | {sum_sepal_length/len(lista):.2f} |')
print(f'|   - width         | {min_sepal_width:2.2f}  | {max_sepal_width:.2f}  | {sum_sepal_width/len(lista):.2f} |')
print(f'| + petal           |       |       |      |                     |')
print(f'|   - length        | {min_petal_length:2.2f}  | {max_petal_length:.2f}  | {sum_petal_length/len(lista):.2f} |')
print(f'|   - width         | {min_petal_width:2.2f}  | {max_petal_width:.2f}  | {sum_petal_width/len(lista):.2f} |')
print(f'+-------------------+--------------------------------------------+')

```
